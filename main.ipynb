```python
conn=mysql.connector.connect(
    host="localhost", 
    user="root", 
    password="desi",
    charset='utf8mb4',

)
    
my_cursor = conn.cursor()
```

```python
my_cursor.execute("CREATE DATABASE IF NOT EXISTS records")
```

```python
conn.database = 'records'
```

```python
#create table for customers
```

```python
my_cursor.execute('''CREATE TABLE IF NOT EXISTS customer_records(
                  CustomerKey INT,
                  Gender VARCHAR(255),
                  Name VARCHAR(255),
                  City TEXT,
                  State_Code VARCHAR(255),
                  State VARCHAR(255),
                  Zip_Code VARCHAR(255),
                  Country VARCHAR(255),
                  Continent VARCHAR(255),
                  Birthday VARCHAR(255),
                  Age INT
                  )''')
print("created successful")
```

```python
# insert values into customers
```

```python
insert_values='''INSERT INTO customer_records(
                  CustomerKey,
                  Gender,
                  Name,
                  City,
                  State_Code,
                  State,
                  Zip_Code,
                  Country,
                  Continent,
                  Birthday,
                  Age)
                  VALUES(%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s)'''
```

```python
customers.shape
```

```python
for index, row in customers.iterrows():
   # print(row['Name'])
    my_cursor.execute(insert_values, (
            row['CustomerKey'],
            row['Gender'],
            row['Name'],
            row['City'],
            row['State Code'],
            row['State'],
            row['Zip Code'],
            row['Country'],
            row['Continent'],
            row['Birthday'],
            row['Age']))

    conn.commit()
```
